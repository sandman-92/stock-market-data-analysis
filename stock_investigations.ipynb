```python
#how often does the stock price go over a set percentage%
import cupy as np
import yfinance as yf
import pandas as pd
import sys
import matplotlib.pyplot as plt
```

```python
#get a list of all companies in SPY from wikipedia
payload = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")
first_table = payload[0]
second_table = payload[1]

first_table.head()
```

```python
companies = first_table["Symbol"].values.tolist()
```

```python
#companies = ['SPY','GOOG', 'AAPL','TSLA']
data = yf.download(companies, start='2016-01-01', end='2022-01-01')
```

```python
data.head()
```

```python
#cell adds the daily percentage difference to the data frame
for company in companies:
    Close = data[('Close', company)]
    Open = data[('Open', company)]
    data = data.copy()
    
    data[('percent_diff',company)] = ((Close - Open)/Close)*100
```

```python
data['percent_diff'].head()
```

```python
#calculate the averages
data[("Averages_percentage_diff", "Mean")] = data["percent_diff"].mean(axis = 1)
data[("Averages_percentage_diff", "Stdev")] = data["percent_diff"].std(axis = 1)
data[("Averages_percentage_diff", "Skew")] = data["percent_diff"].skew(axis = 1)
data[("Averages_percentage_diff", "Kertosis")] = data["percent_diff"].kurt(axis = 1)
data.head()
```

```python
N = 0
limit = 3.0

limit_up_patterns = 0
limit_down_patterns = 0
no_result_patterns = 0


for company in companies:

    stock = data[('percent_diff', company)]
    for i, diff in enumerate(stock[30:]):
        if diff >= limit: #count if the percentage diff is greater than the limit specified
            #save this pattern
            #limit_up_patterns.append(np.array(stock[i-1:i-31]))    
            limit_up_patterns += 1
            N+=1
        elif diff <= -limit:
            #save this pattern
            #limit_down_patterns.append(np.array(stock[i-1:i-31]))
            limit_down_patterns += 1
            N+=1
        else:
            #save pattern
            #no_result_patterns.append(np.array(stock[i-1:i-31]))
            no_result_patterns += 1
            
print("The number of times the stock price diff was greater than ",limit,"was", N)
print("number of up patterns: ",limit_up_patterns) 
print("number of down patterns: ", limit_down_patterns)
print("number of no results:", no_result_patterns)
```
